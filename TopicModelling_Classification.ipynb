```python
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.decomposition import LatentDirichletAllocation

from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.decomposition import NMF

 

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
mc_data = pd.read_excel(r'C:\Users\ashle\Downloads\mckinze_reviews.xlsx')
mc_data.columns
```

```python
mcpros = mc_data['pros']
mcpros =pd.DataFrame(mcpros)
```

```python
mcpros = mcpros.rename(columns = {"pros":"review"})
```

```python
mcpros['review_sentiment']='positive'
mcpros
```

```python
vectorizer = CountVectorizer(max_df=0.8, min_df=4, stop_words='english')

doc_term_matrix = vectorizer.fit_transform(mcpros['review'].values.astype('U'))

doc_term_matrix

 
```

```python
LDA = LatentDirichletAllocation(n_components=5, random_state=35)

LDA.fit(doc_term_matrix)
```

```python
first_topic = LDA.components_[0]

top_topic_words = first_topic.argsort()[-5:]

 

for i in top_topic_words:

   print(vectorizer.get_feature_names()[i])
```

```python
for i,topic in enumerate(LDA.components_):

   print(f'Top 5 words for topic #{i}:')

   print([vectorizer.get_feature_names()[i] for i in topic.argsort()[-5:]])

   print('\n')
```

```python
mccons = mc_data['cons']
mccons =pd.DataFrame(mccons)
mccons = mccons.rename(columns = {"cons":"review"})
mccons['review_sentiment']='negative'
mccons
```

```python
vectorizer = CountVectorizer(max_df=0.8, min_df=3, stop_words='english')

doc_term_matrix = vectorizer.fit_transform(mccons['review'].values.astype('U'))

doc_term_matrix

 
```

```python
LDA = LatentDirichletAllocation(n_components=5, random_state=35)

LDA.fit(doc_term_matrix)
```

```python
for i,topic in enumerate(LDA.components_):

   print(f'Top 5 words for topic #{i}:')

   print([vectorizer.get_feature_names()[i] for i in topic.argsort()[-5:]])

   print('\n')
```

```python
mcreviews = mcpros.append(mccons)
mcreviews
```

```python
import nltk

# download the stopwords dictionary if you have not already

nltk.download('stopwords')

from nltk import word_tokenize, sent_tokenize

from nltk.corpus import stopwords

from nltk.stem import LancasterStemmer, WordNetLemmatizer, PorterStemmer
```

```python
stop = stopwords.words('english')
features = mcreviews['review']

vectorizer = TfidfVectorizer (max_features=70, min_df=3, max_df=0.7, stop_words=stop)

processed_features = vectorizer.fit_transform(features).toarray()
```

```python
labels = mcreviews['review_sentiment']

 

X_train, X_test, y_train, y_test = train_test_split(processed_features, labels, test_size=0.2, random_state=0)

 

text_classifier = RandomForestClassifier(n_estimators=200, random_state=0)

text_classifier.fit(X_train, y_train)
```

```python
predictions = text_classifier.predict(X_test)

 

print(confusion_matrix(y_test,predictions))

print(classification_report(y_test,predictions))

print(accuracy_score(y_test, predictions))
```
